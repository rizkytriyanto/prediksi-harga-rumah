```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('data_rumah.csv', usecols =['bedrooms', 'bathrooms','sqft_living','grade','price','yr_built'])
df.head()
```

Penjelasan setiap kolom:
1. bedrooms = Jumlah kamar tidur
2. bathrooms = Jumlah kamar mandi
3. sqft_living = Luas rumah dalam satuan sqft
4. grade = Grading system dari pemerintah King County US
5. yr_built = Tahun dimana rumah dibangun
6. price = Harga dari rumah ($)

```python
# merubah tipe data dari bathrooms yang semula float menjadi int
df['bathrooms'] = df['bathrooms'].astype('int')
```

```python
# mengganti nilai 33 menjadi 3 di bedrooms
df['bedrooms'] = df['bedrooms'].replace(33,3)
```

```python
df.corr().style.background_gradient().set_precision(2)
```

```python
x = df.drop(columns='price')
y = df['price']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=4)

lin_reg = LinearRegression()

lin_reg.fit(x_train, y_train)

print(lin_reg.coef_)
print(lin_reg.intercept_)
```

```python
coef_dict = {
    'features': x.columns,
    'coef_value':lin_reg.coef_
}
coef = pd.DataFrame(coef_dict, columns=['features', 'coef_value'])
coef
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
lin_reg.score(x_test, y_test)
```
